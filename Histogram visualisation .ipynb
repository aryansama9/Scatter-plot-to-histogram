```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("assessment_3_dataset_1.csv")

x = dataset.iloc[:, -2]
y = dataset.iloc[:, -1]

frame, axis = plt.subplots(2, 2, gridspec_kw = {"width_ratios": [4,1], "height_ratios": [1, 4]}, 
    figsize = (10, 8), 
    sharex = "col",
    sharey = "row") 

#sharex, sharey, gridspec_kw, "col", "row", "width_ratios" etc are a part of matplotlib for sizing the frames to a reasonable looking size

scatterplot = axis[1, 0]
scatterplot.scatter(x, y)
scatterplot.set_xlabel("Z axis")
scatterplot.set_ylabel("Y axis")
scatterplot.set_title("Scatter plot")
y_histogram = axis[1, 1]
y_histogram.hist(y, orientation = "horizontal", color = "red", bins = 200) #200 red bars tilted horizontally
#To arrange the y axis's histogram to tilt I had to add orientation
y_histogram.set_xlabel("Y axis's histogram") #add the name to the x axis because it is telling the y axis same goes for the y axis for z

x_histogram = axis[0, 0]
x_histogram.hist(x, color = "purple", bins = 200) #200 purple bars
x_histogram.set_ylabel("Z axis's histogram")
#In order to get a proper amount I set bins = 200 meaning the amount of bars for both columns as each column has 200 in reference to the plot 0 to 100 and 0 to -100
plt.show()
```
